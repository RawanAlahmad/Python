```python
import pandas as pd

# Load the CSV file into a DataFrame
file_path = r"C:\Users\Rawana\Desktop\NoteBook Module3\Ventes_Samsung.csv"
df = pd.read_csv(file_path)

# afficher les premières lignes
df.head()
```

<h3 style=" color:black ; font-family:  'Comic Sans MS', sans-serif; font-size: 19px; ">2. Afficher le nombre total de ventes enregistrées.</h3>

```python
# Calculer le nombre total de ventes enregistrées.
total_sales = len(df)
print("le nombre total de ventes enregistrées.:", total_sales)
```

<h3 style=" color:black ; font-family:  'Comic Sans MS', sans-serif; font-size: 19px; ">3. Calculer et afficher le montant total des ventes.</h3>

```python
# Calculer le montant total des ventes 
montant_total = df['Montant_Total'].sum()

# Afficher le montant total des ventes 
print("Le montant total des ventes :",montant_total)
```

<h1 style="color:blue ; font-family: 'Comic Sans MS', sans-serif; font-size: 24px;text-align: center; text-shadow: 1px 1px #ff69b4 ">Partie 2 : Analyse des ventes ___________________________________________________</h1>

<h3 style=" color:black ; font-family:  'Comic Sans MS', sans-serif; font-size: 19px; ">1. Identifier le produit le plus vendu (en quantité).</h3>

```python
# Regrouper par ID de produit et additionner les quantités vendues
product_sales = df.groupby('ID_Produit')['Quantité_Vendue'].sum()

#  Identifiez le produit avec la plus grande quantité vendue
most_sold_product = product_sales.idxmax()
most_sold_quantity = product_sales.max()

# Afficher le produit le plus vendu et la quantité vendue
print(f"Le produit le plus vendu t: {most_sold_product}")
print(f"La quantité vendue: {most_sold_quantity}")
```

<h3 style=" color:black ; font-family:  'Comic Sans MS', sans-serif; font-size: 19px; ">2. Calculer le revenu moyen par vente.
.</h3>

```python
# Calculer le montant total des ventes
montant_total_ventes = df['Montant_Total'].sum()

# Calculer le nombre total de ventes
nombre_total_ventes = len(df)

# Calculer le revenu moyen par vente
revenu_moyen_par_vente = montant_total_ventes / nombre_total_ventes

# Afficher le revenu moyen par vente
print(f" le revenu moyen par ventee: {revenu_moyen_par_vente :.2f}")
```

<h3 style=" color:black ; font-family:  'Comic Sans MS', sans-serif; font-size: 19px; ">3. Déterminer le pays avec le plus grand nombre de ventes.</h3>


```python
# Regrouper par pays et compter le nombre de ventes
ventes_par_pays = df['Pays_Vente'].value_counts()

# Identifiez le pays avec le plus grand nombre de ventes
most_sales_country = ventes_par_pays.idxmax()
most_sales_count = ventes_par_pays.max()

# Afficher le pays avec le plus grand nombre de ventes et le décompte
print(f"le pays avec le plus grand nombre de ventes : {most_sales_country}")
print(f"Nombre de ventes : {most_sales_count}")
```

<h1 style="color:blue ; font-family: 'Comic Sans MS', sans-serif; font-size: 24px;text-align: center; text-shadow: 1px 1px #ff69b4 ">Partie 3 : Manipulation des données
 ___________________________________________________</h1>

<h3 style=" color:black ; font-family:  'Comic Sans MS', sans-serif; font-size: 19px; ">1. Ajouter une colonne 'Montant_Moyen_Par_Produit' qui représente le montant moyen 
dépensé pour chaque produit vendu</h3>


```python
# Calculez le montant total et la quantité totale vendue pour chaque produit
product_totals = df.groupby('ID_Produit').agg({'Montant_Total': 'sum', 'Quantité_Vendue': 'sum'}).reset_index()

# Calculer le montant moyen dépensé par produit
product_totals['Montant_Moyen_Par_Produit'] = product_totals['Montant_Total'] / product_totals['Quantité_Vendue']

# Fusionner ces informations dans le DataFrame d'origine
df = pd.merge(df, product_totals[['ID_Produit', 'Montant_Moyen_Par_Produit']], on='ID_Produit', how='left')

# Supprimez les colonnes en double si elles existent (facultatif, juste pour le nettoyage)
df = df.loc[:,~df.columns.duplicated()]

# Afficher le DataFrame mis à jour avec la nouvelle colonne
df.head()
```

<h3 style=" color:black ; font-family:  'Comic Sans MS', sans-serif; font-size: 19px; ">2. Filtrer les données pour inclure uniquement les ventes réalisées en France et afficher les 
résultats.</h3>

```python
# Filtrer les données pour inclure uniquement les ventes réalisées en France 
sales_france = df[df['Pays_Vente'] == 'France']

# Afficher les résultats
sales_france
```

<h3 style=" color:black ; font-family:  'Comic Sans MS', sans-serif; font-size: 19px; ">3. Trier les données par 'Score_Satisfaction' de manière décroissante et afficher les 10 
premières lignes.</h3>

```python
# Trier les données par 'Score_Satisfaction' de manière décroissante 
sorted_data = df.sort_values(by='Score_Satisfaction', ascending=False)

#  Afficher les 10 premières lignes.
top_10_satisfaction = sorted_data.head(10)

top_10_satisfaction
```

<h3 style=" color:black ; font-family:  'Comic Sans MS', sans-serif; font-size: 19px; ">1. Créer un histogramme représentant la distribution du 'Montant_Total' des ventes.</h3>

```python
import matplotlib.pyplot as plt


#  Créer un histogramme représentant la distribution du 'Montant_Total' des ventes. 
plt.figure(figsize=(10, 6))
plt.hist(df['Montant_Total'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Distribution of Montant_Total')
plt.xlabel('Montant_Total')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

<h3 style=" color:black ; font-family:  'Comic Sans MS', sans-serif; font-size: 19px; ">2. Générer un graphique en barres montrant le nombre de ventes par 'Canal_Achat'.</h3>

```python
# Générer un graphique en barres montrant le nombre de ventes par 'Canal_Achat'.

sales_by_channel = df['Canal_Achat'].value_counts()

# Générer le graphique à barres
plt.figure(figsize=(10, 6))
sales_by_channel.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Sales by Canal_Achat')
plt.xlabel('Canal_Achat')
plt.ylabel('Number of Sales')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

<h3 style=" color:black ; font-family:  'Comic Sans MS', sans-serif; font-size: 19px; ">3. Visualiser la relation entre 'Délai_Livraison_Jours' et 'Score_Satisfaction' à l'aide d'un 
nuage de points</h3>

```python
# Générer un nuage de points
plt.figure(figsize=(10, 6))
plt.scatter(df['Délai_Livraison_Jours'], df['Score_Satisfaction'], alpha=0.6, edgecolors='w', s=80)
plt.title('La relation entre Délai_Livraison_Jours et Score_Satisfaction')
plt.xlabel('Délai_Livraison_Jours')
plt.ylabel('Score_Satisfaction')
plt.grid(True)
plt.show()
```
