```python
# Re-import necessary libraries and reload the data in case the previous context was lost
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = (r"C:\Users\Rawana\Desktop\NoteBook Module3\Publicite_Cosmetiques_Data.xlsx")
data = pd.read_excel(file_path, sheet_name='Données')

# Visualize missing values using pandas and matplotlib
plt.figure(figsize=(10, 6))
plt.imshow(data.isnull(), aspect='auto', cmap='viridis', interpolation='nearest')
plt.colorbar(label='Missing Data')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.title('Missing Values in DataFrame')
plt.xticks(range(len(data.columns)), data.columns, rotation=90)
plt.show()

# Calculate the number of missing values for each column
missing_values = data.isnull().sum()

# Filter columns with missing values
columns_with_missing_values = missing_values[missing_values > 0]
columns_with_missing_values

```

La visualisation indique qu'il n'y a aucune valeur manquante dans aucune des colonnes du DataFrame, car la série montrant le nombre de valeurs manquantes pour chaque colonne est vide.

- Remplacer les valeurs manquantes dans la colonne "Taux de Conversion (%)" 
par la valeur moyenne de cette colonne. (Indice : Utilisez fillna() avec mean()).

```python
# Calculate the mean of the 'Taux de Conversion (%)' column
conversion_rate_mean = data['Taux de Conversion (%)'].mean()

# Replace missing values in 'Taux de Conversion (%)' column with the mean
data['Taux de Conversion (%)'].fillna(conversion_rate_mean, inplace=True)

# Verify that there are no missing values in 'Taux de Conversion (%)' column
data['Taux de Conversion (%)'].isnull().sum()

```

# Exercice 4
1. Écrire une fonction calculer_ROI qui prend en entrée le budget de la campagne 
et les ventes générées, et retourne le Retour sur Investissement (ROI)

```python
def calculer_ROI(budget, ventes):
    return (ventes - budget) / budget
```

- 2. Appliquer cette fonction à chaque ligne du DataFrame pour créer une nouvelle 
colonne "ROI".

```python
data['ROI'] = data.apply(lambda row: calculer_ROI(row['Budget de la Campagne (€)'], row['Ventes Générées (€)']), axis=1)
```

- 3. Créer une fonction classer_segment qui prend l'âge du segment et retourne 
"Jeune", "Adulte" ou "Senior" selon l'âge. Appliquer cette fonction à la colonne 
"Segmentation du Public - Âge" pour créer une nouvelle colonne "Classe d'Âge".

```python
def classer_segment(age_segment):
    if age_segment in ['18-24']:
        return 'Jeune'
    elif age_segment in ['25-34', '35-44']:
        return 'Adulte'
    else:
        return 'Senior'
```

```python
data['Classe d\'Âge'] = data['Segmentation du Public - Âge'].apply(classer_segment)
```
