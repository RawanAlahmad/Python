<h2 style="color:black ; font-family: 'Comic Sans MS', sans-serif; font-size: 24px; text-shadow: 1px 1px #ff69b4 "> 4.Fonctions: </h2>
<h2 style=" color:#9932cc ; font-family:  'Comic Sans MS', sans-serif; font-size: 19px; " >
- Écrivez une fonction qui prend en entrée le nom d'un programme etrenvoie son 
audience et sa part de marché</h2>


```python
def get_audience_and_market_share(program_name):
    # Search for the program in the dataframe
    program_data = df[df['Nom_programme'] == program_name]
    
    if not program_data.empty:
        audience = program_data['Audience'].values[0]
        market_share = program_data['Part_marché'].values[0]
        return audience, market_share
    else:
        return None, None
```

```python
get_audience_and_market_share("Koh-Lanta")
```

```python
get_audience_and_market_share("Miss France")
```

```python
example_programs = ["Jacky Show", "Face à la Une", "La Chanson secrète"]

results = {program: get_audience_and_market_share(program) for program in example_programs}

results
```

<h2 style="color:#ff1493; font-family:  'Comic Sans MS', sans-serif; font-size: 24px; text-shadow: 1px 1px #ff69b4; ">Partie 3 : Analyse </h>
<h2 style="color:black ; font-family: 'Comic Sans MS', sans-serif; font-size: 24px; text-shadow: 1px 1px #ff69b4 "> 1. Statistiques Descriptives :</h2>
<h2 style=" color:#9932cc ; font-family:  'Comic Sans MS', sans-serif; font-size: 19px; " >
Calculez les statistiques descriptives de base (moyenne, médiane, écart-type) pour 
l'audience et les revenus publicitaires.</h2>


```python
# Calculer des statistiques descriptives pour 'Audience' et 'Revenus_publicitaires'
audience_stats = df['Audience'].describe()
revenus_stats = df['Revenus_publicitaires'].describe()

# Extraire la moyenne, la médiane et l'écart type pour les deux colonnes
audience_mean = audience_stats['mean']
audience_median = df['Audience'].median()
audience_std = audience_stats['std']

revenus_mean = revenus_stats['mean']
revenus_median = df['Revenus_publicitaires'].median()
revenus_std = revenus_stats['std']


# Créez une trame de données récapitulative pour une visualisation plus facile
summary_stats = pd.DataFrame({
    'Statistic': ['Mean', 'Median', 'Standard Deviation'],
    'Audience': [audience_mean, audience_median, audience_std],
    'Revenus Publicitaires': [revenus_mean, revenus_median, revenus_std]
})

summary_stats
```
